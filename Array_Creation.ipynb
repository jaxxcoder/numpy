```python
import numpy as np 
```

## Creating an array using numpy's array() function

```python
array_elements=[[2,9,7],[7,9,2]]
myArray=np.array(array_elements)
```

```python
myArray
```

## Creating an array using numpy's zeros() function

```python
myArray2=np.zeros((3,5))
```

```python
myArray2
```

*It creates an array according to the user's input with initial value 0

## Creating an array using numpy's empty() function

```python
myArray3=np.empty((2,4))
```

```python
myArray3
```

## Creating an array using numpy's arange() function

```python
myArray5=np.arange(20)
```

```python
myArray5
```
